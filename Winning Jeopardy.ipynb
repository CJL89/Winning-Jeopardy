```python
import re
import numpy as np
import pandas as pd

print('Installed all dependecies!')
```

```python
# Saving the sataset in a variable and displaying the top rows:

jeopardy = pd.read_csv('jeopardy.csv')
jeopardy.head(10)
```

```python
# Printing the columns of the dataset:

jeopardy.columns
```

```python
# Eliminiting whitespace before the column's name:

jeopardy.columns = ['Show Number', 'Air Date', 'Round', 'Category', 'Value', 'Question', 'Answer']
jeopardy.columns
```

```python
# Creating function that normalizes the text dataset:

def normalize(text):
    text = str(text)
    text = text.lower()
    text = re.sub("[^A-Za-z0-9\s]", "", text)
    return text

print('Function ready to use!')
```

```python
# Normalizing the Question and Answers columns:

jeopardy["clean_question"] = jeopardy["Question"].apply(normalize)
jeopardy["clean_answer"] = jeopardy["Answer"].apply(normalize)

jeopardy.head(10)
```

```python
# Creating function that normalizes the numeric dataset:

def normalize_values(value):
    value = re.sub("[^A-Za-z0-9\s]", '', value)
    try:
        value = int(value)
    except Exception:
        value = 0
    return value

print('Function ready to use!')
```

```python
jeopardy['clean_value'] = jeopardy['Value'].apply(normalize_values)
jeopardy.head(10)
```

```python
jeopardy['clean_air_date'] = pd.to_datetime(jeopardy['Air Date'])
jeopardy.head(10)
```
